```python
import torch
from torch_geometric.data import Data
```

```python
# `Data` is fundamental graph data structure in PyG with params
#   edge_index: tuple of source and destination node indices for each edge
#   x: node features - input attributes for each node
#   y: node labels - target (ground truth) for each node
#   train_mask = boolean tensor used to section training data (1) from other (0)

edge_index = torch.tensor([[0, 1, 2, 1],  # sources
                           [1, 2, 0, 2]], dtype=torch.long)  # destinations

x = torch.tensor([[-1], [0], [1]], dtype=torch.float)

data = Data(x=x, edge_index=edge_index)
# this graph has 3 nodes 4 directional edges total
print(data)
```

```python
# you can also represent edge_index as a list of [source, dest] index tuples

edge_index = torch.tensor([[0, 1],
                           [1, 0],
                           [1, 2],
                           [2, 1]], dtype=torch.long)

x = torch.tensor([[-1], [0], [1]], dtype=torch.float)

data = Data(x=x, edge_index=edge_index.t().contiguous())  # .t() = transpose
print(data)
```

```python
# check graph construction with

is_valid = data.validate(raise_on_error=True)
print(f"Data object is valid: {is_valid}")
```

```python
# useful utility functions to investigate graphs

print(data.keys())
print(data['x'])

for key, item in data:
    print(f'{key} found in data')

print('edge_attr' in data)
print('x' in data)

print("# of nodes: ", data.num_nodes, "   # of edges: ", data.num_edges)
print("# of features: ", data.num_node_features)

print(data.has_isolated_nodes())
print(data.has_self_loops())
```

```python
# move between CPU and GPU on my M1 chip macbook

new_device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

choice = input(f"transfer to {new_device}? (y/n): ").strip().lower()

if choice == "y":
    if data.x.device.type == new_device.type:
        print(f"Already on {new_device}!")
    else:
        print(f"Transferring model to {new_device}...")
        data = data.to(new_device)
        x = x.to(new_device)
elif choice == "n":
    print("Skipping.")
```
